# Challenge 1

#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
#subplot(nrows, ncols, index, **kwargs)
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,z)
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
# subplot(pos, **kwargs)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('First plot')

fig, ax = plt.subplots()
ax.plot(x, z)
ax.set_title('Second plot')
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
#f, axs = plt.subplots(2,2,figsize=(15,15))

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, y)
ax.set_title('First plot')

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, z)
ax.set_title('Second plot')
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('First plot')

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, z)
ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_title('Second plot')
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
fig, ax = plt.subplots(1, 2, figsize=(10,3))
ax[0].plot(x, x**2, x, np.exp(x))
ax[0].set_title("normal scale")

ax[1].plot(x, x**2, x, np.exp(x))
ax[1].set_yscale('log')
ax[1].set_title("log scale")
```
