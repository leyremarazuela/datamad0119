```python
import requests
from bs4 import BeautifulSoup
```

```python
def getNoticia(url):
    html = requests.get(url).content
    soup = BeautifulSoup(html, "lxml")                                           
    noticia = {
       "titular": soup.find_all('h1')[0].text 
    }
    #noticia = "titular": soup.find_all('h1')[0].text
    return noticia

display(getNoticia('https://www.reuters.com/article/us-shazam-m-a-apple-eu/eu-clears-apples-purchase-of-shazam-idUSKCN1LM1TZ'))
display(getNoticia('https://www.reuters.com/article/us-britain-labour-technology/well-take-on-social-media-giants-uks-opposition-labour-party-says-idUSKCN1PV0PL'))
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
```

```python
table = soup.find_all('table',{'class':'sortable wikitable'})[0]
rows = table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
rows
```

```python
import pandas as pd 
df = pd.DataFrame(rows[1:],columns=rows[0])
```

```python
df = df.sort_values(ascending=False, by="Life expectancy")
df = df.reset_index(drop=True)
df["Life expectancy"] = df["Life expectancy"].astype(int)
df
```

```python
df["多Mola?"] = pd.cut(df["Life expectancy"], bins=[df["Life expectancy"].min()-1,80,df["Life expectancy"].max()+1], labels=["Nomola","Mola"])
```

```python
df.groupby("多Mola?").agg({"Life expectancy":"mean"})
```
