#### 6. Import the posts table. 

```python
posts_table = pd.read_sql_query('SELECT * FROM posts', engine)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts_table=posts_table.rename(columns={"Id": "postId", "OwnerUserId": "userId"})
posts_table.head()
```

```python
users_df=users_table[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
posts_df=posts_table[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
print(users_df)
print(posts_df)
```

```python
posts_users_df=users_df.merge(posts_df)
posts_users_df
```
